```python
import pandas as pd
from pathlib import Path
```

# Column Manipulation

```python
csvpath = Path("../Resources/customers.csv")
```

```python
customer_dataframe = pd.read_csv(csvpath)
customer_dataframe.head()
```

### Replace Columns

```python
customer_dataframe.columns
```

```python
# Rewrite the column names
columns = ["Full Name", "Email", "Address", "Zip Code", "Credit Card Number", "Account Balance"]
customer_dataframe.columns = columns
customer_dataframe.head()
```

### Rename Columns

```python
customer_dataframe=customer_dataframe.rename(columns={
    "Full Name": "full_name",
    "Credit Card Number": "credit_card_number"
})

#OR set }, inplace = True

customer_dataframe.head()
```

```python
#Selecting specific columns, must use list within list for more than one column name
customer_dataframe[['full_name', 'Email']]
```

### Reorder Columns

```python
customer_dataframe = customer_dataframe[['credit_card_number', 'Account Balance', 'full_name', 'Email', 'Address', 'Zip Code']]
customer_dataframe.head()
```

### Create Columns

```python
customer_dataframe["Balance (1k)"] = customer_dataframe["Account Balance"] / 1000
customer_dataframe.head()
customer_dataframe["Double Balance"] = customer_dataframe ["Account Balance"] * 2
customer_dataframe.head()
```

### Split Columns

```python
names = customer_dataframe["full_name"].str.split(" ", expand=True)
names.head()
```

```python
names["first_name"] = names[0]
names["last_name"] = names[1]
names.drop(columns = [0,1])
```

### Delete Columns

```python
names = customer_dataframe = customer_dataframe.drop(columns=["Double Balance"])
names.head()
```
